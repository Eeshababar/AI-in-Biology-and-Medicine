```python
#K=1
import numpy as np
np.random.seed(0) #Using random seed constant value so we get same split of data everytime
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
iris = np.genfromtxt(fname = '/content/iris.txt', delimiter=None)
X = iris[:, 0:2] #Using first two feature
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y) #Shuffle the data
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.8)  #Dividing data into train, validate 80,20 percent
print(Xtr.shape)
print(Xva.shape)
# Creating an classifier.
knn = ml.knn.knnClassify()
# Training the classifier.
knn.train(Xtr, Ytr, K=1)
# Making predictions
YvaHat = knn.predict(Xva)
print(YvaHat)
trainerror=knn.err(Xtr, Ytr)
print(trainerror)
validateerror =knn.err(Xva,Yva)
print(validateerror)
ml.plotClassify2D(knn, Xtr, Ytr);
```

##Q2.2

```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
trainerror = [] #Create an empty list named as trainerror
validateerror = []  #Create an empty list validate error
iris = np.genfromtxt(fname = '/content/iris.txt', delimiter=None)
X = iris[:, 0:2]  #first two features only
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.75)
print(Xtr.shape)
print(Xva.shape)
K=[1,2,5,10,50,100,200];
for k in K:
    # Creating an classifier.
    knn = ml.knn.knnClassify()
    # Training the classifier.
    knn.train(Xtr, Ytr, k)

    # Making predictions
    YvaHat = knn.predict(Xva)
    print(YvaHat)
    knn.err(Xtr, Ytr) #get training error
    knn.err(Xva,Yva)  #get validation error
    trainerror.append(knn.err(Xtr, Ytr)) #After each iteration append trainerror to list created above
    validateerror.append(knn.err(Xva,Yva)) #After each iteration append validateerror to list created above
print(trainerror)
print(validateerror)
plt.semilogx(K, trainerror, 'r-', lw=3, label='Training')
plt.semilogx(K, validateerror, 'g-', lw=3, label='Validation')
plt.legend()
```

K = 50 recommended here as validation error is minimum at K = 50 and increases after that

```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
trainerror = [] #Create an empty list named as trainerror
validateerror = []  #Create an empty list validate error
iris = np.genfromtxt(fname = '/content/iris.txt', delimiter=None)
X = iris[:, 0:4]
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.75)
print(Xtr.shape)
print(Xva.shape)
K=[1,2,5,10,50,100,200];
for k in K:
    # Creating an classifier.
    knn = ml.knn.knnClassify()
    # Training the classifier.
    knn.train(Xtr, Ytr, k)

    # Making predictions
    YvaHat = knn.predict(Xva)
    print(YvaHat)
    knn.err(Xtr, Ytr) #get training error
    knn.err(Xva,Yva)  #get validation error
    trainerror.append(knn.err(Xtr, Ytr)) #After each iteration append trainerror to list created above
    validateerror.append(knn.err(Xva,Yva)) #After each iteration append validateerror to list created above
print(trainerror)
print(validateerror)
plt.semilogx(K, trainerror, 'r-', lw=3, label='Training')
plt.semilogx(K, validateerror, 'g-', lw=3, label='Validation')
plt.legend()
```

Error function is more smooth when using all features and the error values are less in this case as compared when we used only two features.
For first graph with two features my recommendation was K = 50. In this graph (using all four features) the error value is small for two values of K, K = 10 and K = 50. I will recommend K = 50 in this case too because by using K = 50 we will have more smooth boundry and less complexity
